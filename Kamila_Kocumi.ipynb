SECTION N2: DATABASES VARIABLE REORDERING AND MERGING

```python
# Before merging the dataframes it makes sense to check if the dataframes are made up by the very same columns
# and if those colums appear in the very same order

# List of DataFrames
dataframes = [vehicules_2017, vehicules_2018, vehicules_2019, vehicules_2020, vehicules_2021, vehicules_2022]

# Check if all DataFrames have the same columns in the same order
same_columns = all(df.columns.tolist() == dataframes[0].columns.tolist() for df in dataframes[1:])

if same_columns:
    print("All DataFrames have the same columns in the same order.")
else:
    print("DataFrames have different columns or different column orders.")
```

```python
# The title of each variables are renamed and standardize to avoid spelling issue.
# All databases will have the variable names reordered alphabetically.

import pandas as pd

# Function to normalize columns across dataframes
def normalize_columns(dataframes):
    # Identify all unique columns across all dataframes
    all_columns = set(column for df in dataframes for column in df.columns)

    # Sort the columns alphabetically
    standard_columns_order = sorted(list(all_columns))

    # Function to reorder columns in a dataframe based on the standard order
    def reorder_columns(df):
        # Add missing columns
        for column in standard_columns_order:
            if column not in df.columns:
                df[column] = None

        # Reorder columns based on the standard order
        df = df[standard_columns_order]
        return df

    # Apply the reorder_columns function to each dataframe
    dataframes_normalized = [reorder_columns(df) for df in dataframes]

    return dataframes_normalized, standard_columns_order

# Normalize columns across dataframes
dataframes_normalized, standard_columns_order = normalize_columns(dataframes)

# Display the updated column names for each normalized dataframe
for i, df in enumerate(dataframes_normalized):
    print(f"Columns for vehicules_{2017 + i}: {df.columns.tolist()}")

# Display the standard columns order
print("Standard Columns Order:", standard_columns_order)
```

```python
# Assuming dataframes_normalized is the list of normalized dataframes from the previous code.
# The databases gets sum up into a single one.
# The databases will be in ordered with the oldest year's information on the top, following the order given by the list "dataframes_normalized".

# Concatenate the dataframes along the rows
merged_dataframe = pd.concat(dataframes_normalized, ignore_index=True)

# Assignn a useful name to the merged database. Among the suggested ones for this project:
# "Caracteristiques2017-2022", "Lieux2017-2022", "Usagers2017-2022" or "Vehicules2017-2022".
Vehicules2017_2022 = merged_dataframe

# Display the first few rows of the merged dataframe
print("Merged DataFrame:")
print(merged_dataframe.head())

# Display the column names of the merged dataframe
print("\nColumns of the Merged DataFrame:")
print(merged_dataframe.columns.tolist())
```
